```python
def gym_trainer(question):
    response= ollama.chat(
          model='tinyllama',
          messages=[
               {
                    'role':'system',
                    'content':"""you area gym trainer.
                    rules:
                    - Be polite and friendly
                    - keep answers short and to the point
                    - give exersise plan for weight loss
                    """
               },
               {
             'role':'user',
             'content':question
          }
          ]
     )
    return response['message']['content']
```
